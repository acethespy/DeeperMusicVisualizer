```python
import sys, os
import pygame
import numpy as np

from pydub import AudioSegment
from pydub.playback import play

sys.path.append(SYM_PATH)
sys.path.append(os.path.join(SYM_PATH,'osu_beatmap_parser'))

from osureader.reader import BeatmapParser
from osureader.beatmap import Beatmap
from osureader.objects import HitObjectType
from data_manager.data_manager import DataManager as DM

import IPython
```

```python
beatmap2 = None

folder = os.path.join(SYM_PATH, 'data/beatmaps')
beatmaps, music_file = DM.get_beatmap(folder, True, True)

beatmap = beatmaps[0]
beatmap2 = beatmaps[1]

print(music_file)
```

```python
print(beatmap.general_settings.audio_file_name)
print(beatmap.general_settings.audio_lead_in)
print(beatmap.general_settings.preview_time)
print(beatmap.general_settings.countdown)
for i in range(10):
    hit_object = beatmap.hit_objects[i]
    print(f'{i}: {hit_object.type} {hit_object.time}')
# Time is in milliseconds
# print(len(beatmap.hit_object))
print(len(beatmap.timing_objects))
for timing_object in beatmap.timing_objects:
    if timing_object.uninherited == 1:
        print(f'{timing_object.time} {timing_object.beat_length}')
```

```python
hit_objects = beatmap.hit_objects.copy();
timing_objects = beatmap.timing_objects.copy();

hit_times = [hit_object.time for hit_object in hit_objects]

song = AudioSegment.from_mp3(music_file)
length = len(song)
track = AudioSegment.silent(duration=length)
hat = AudioSegment.from_mp3(os.path.join(SYM_PATH, "visualizer/high_hat.mp3"))


for time in hit_times:
    track = track.overlay(hat, position=time)


mixed = song.overlay(track)

out_dir = os.path.join(SYM_PATH, 'data/beatmaptomp3_results')
track.export(os.path.join(out_dir, "boombayah_hit.wav"), format="wav")
mixed.export(os.path.join(out_dir, "boombayah_mixed.wav"), format="wav")
```
