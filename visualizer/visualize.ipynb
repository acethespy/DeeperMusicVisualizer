#Generate scripts for experiements

```python
song_names = ['_57/Faded']
output_suffix = ['_128_50epochs_hardcoded_alltiming', '_128_50epochs_hardcoded_singletiming']


for name in song_names:
  osu_file = os.path.join(SYM_PATH, f"data/beatmaps{name}/beatmap.osu")
  music_file = os.path.join(SYM_PATH, f'data/beatmaps{name}/audio.mp3')
  result_file = os.path.join(SYM_PATH, f'results/results{name}_osu.mp4')
  if not os.path.exists(result_file):
    print(result_file, "doesn't exist so generating video")
    L = LucidSonicDream(song = music_file,
                        pulse_audio = music_file,
                        class_audio = music_file,
                        contrast_audio = music_file,
                        style = style,
                        osu_file=osu_file)
    L.hallucinate(file_name = result_file,
                  resolution = 720,
                  start = 0,
                  duration=60,
                  pulse_react = 0.05, 
                  motion_react = 0.001)
  else:
    print(result_file, "exists already")
  for suffix in output_suffix:
    osu_file = os.path.join(SYM_PATH, f"data/beatmaps{name}/output{suffix}.npy")
    music_file = os.path.join(SYM_PATH, f'data/beatmaps{name}/audio.mp3')
    result_file = os.path.join(SYM_PATH, f'results/results{name}{suffix}.mp4')

    L = LucidSonicDream(song = music_file,
                        pulse_audio = music_file,
                        class_audio = music_file,
                        contrast_audio = music_file,
                        style = style,
                        osu_file=osu_file)
    L.hallucinate(file_name = result_file,
                  resolution = 720,
                  start = 0,
                  duration=60,
                  pulse_react = 0.05, 
                  motion_react = 0.001)
```

```python
# song_names = ['Stargazing', 'Moments', 'AnotherLife', 'LightSH', 'KuagaLostTime', 'PretenderAoki', 'ItAintMe', 'WakeUpExcisionSK']
song_names = ['Firework']
output_suffix = ['_128_50epochs_essentia']
existing_songs = []

for name in song_names:
  for suffix in output_suffix:
    osu_file = os.path.join(SYM_PATH, f"data/songs/{name}/output{suffix}.npy")
    music_file = os.path.join(SYM_PATH, f'data/songs/{name}/audio.mp3')
    result_file = os.path.join(SYM_PATH, f'results/songs/{name}{suffix}_afxtvid.mp4')

    if not os.path.exists(result_file):
      L = LucidSonicDream(song = music_file,
                          pulse_audio = music_file,
                          class_audio = music_file,
                          contrast_audio = music_file,
                          style = style,
                          osu_file=osu_file)
      L.hallucinate(file_name = result_file,
                    resolution = 720,
                    start = 80,
                    duration=60,
                    batch_size = 10,
                    pulse_react = 0.05, 
                    motion_react = 0.001)
    else:
      existing_songs.append(name)
      print("Skipping ", name)
    print("Existing songs:", existing_songs)
```
