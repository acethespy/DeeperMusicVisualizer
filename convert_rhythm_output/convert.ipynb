```python
file_path = os.path.join(SYM_PATH, 'convert_rhythm_output/rhythm_data_faded_128_50epochs_hardcoded_alltiming.npz')
data = np.load(file_path)
```

```python
objs = data['objs']
predictions = data['predictions']
ticks = data['ticks']
timestamps = data['timestamps']
is_slider = data['is_slider']
is_spinner = data['is_spinner']
is_note_end = data['is_note_end']
sv = data['sv']
slider_ticks = data['slider_ticks']
dist_multiplier = data['dist_multiplier']
```

```python
result = []
sliding = False;
for i, pred in enumerate(predictions):
  if pred[2]==1: #slider
    slide_ticks = slider_ticks[i]
    if i == len(predictions) - 1:
      slide_ms = 1000
    elif ticks[i+1] - ticks[i] > 0:
      ms_gap = timestamps[i+1] - timestamps[i]
      slide_ms = slide_ticks / (ticks[i+1] - ticks[i]) * ms_gap
    else:
      ms_gap = timestamps[i+1] - timestamps[i]
      slide_ms = slide_ticks / (4) * ms_gap
    if sliding==False or i==len(predictions) - 1:
      result.append([timestamps[i], 1, slide_ms])
    else:
      result[-1][2] += slide_ms
    if i!=len(predictions) - 1 and ticks[i+1] - ticks[i] == slide_ticks:
      print('continuing slide')
      sliding = True;
  elif pred[1]==1: #note
    sliding=False;
    result.append([timestamps[i], 0, 0])
  else:
    print("Error")
```

```python
print(result)
```

```python
np.save('/content/DeepMusicVisualizer/data/beatmaps_57/Faded/output_128_50epochs_hardcoded_alltiming.npy', result)
```
