## Documentation
Iterates through each .zip/.osz file in the directory /data/osz.

For each file, the user is prompted for the name of the resulting folder for the content of the zipped file which will be located at /data/beatmaps.

```python
import sys, os
import zipfile

beatmap_folder = os.path.join(SYM_PATH, 'data/beatmaps')
osz_folder = os.path.join(SYM_PATH, 'data/osz')

for filename in os.listdir(osz_folder):
  print(filename)
  if filename.endswith(".osz") or filename.endswith(".zip"):
    print(f'File name: {filename}')
    print(f'Enter folder name: ')
    name = input()
    result_path = os.path.join(beatmap_folder, name)
    og_file_path = os.path.join(osz_folder, filename)
    if filename.endswith(".osz"):
      pre, ext = os.path.splitext(osz_file_path)
      zip_file_path = pre + '.zip'
      os.rename(og_file_path, zip_file_path)
    else:
      zip_file_path = og_file_path
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
      zip_ref.extractall(result_path)
    if os.path.exists(og_file_path):
      os.remove(og_file_path)
    else:
      print(f"Error deleting {osz_file_path} file doesn't exist")
  else:
    print(f'SKIP osz/zip extension not found: {filename}')
```
