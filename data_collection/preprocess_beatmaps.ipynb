```python
beatmap_dir = os.path.join(SYM_PATH, 'data/beatmaps')
result_dir = os.path.join(SYM_PATH, 'data/beatmaps_illenium')

beatmap = None
music_path = None
parser = BeatmapParser()

results = [bname for bname in os.listdir(result_dir) if os.path.isdir(os.path.join(result_dir, bname))]
beatmaps = [{'name': bname, 'dir_path': os.path.join(beatmap_dir, bname)} for bname in os.listdir(beatmap_dir) if os.path.isdir(os.path.join(beatmap_dir, bname)) and bname not in results]

print(results)

print('Available Songs: ')
for i in range(len(beatmaps)):
  print(f'{i:02d}) {beatmaps[i]["name"]}')

for i in range(len(beatmaps)):
  choice = beatmaps[i]
  options = []
  for filename in os.listdir(choice['dir_path']):
    if filename.endswith(".osu"):
      start = filename.rfind('[')
      end = filename.rfind(']')
      if start != -1 and end != -1:
        name = filename[start+1:end]
      else:
        name = filename
      bm=Beatmap(parser.parse(os.path.join(choice['dir_path'], filename)))
      options.append({'name': name, 'filename': filename, 'difficulty': bm.difficult_settings.overall_difficulty, 'beatmap': bm, 'artist':bm.meta_data.artist})
  assert(len(options) == 0, f'no .osu files found for {choice["name"]}')
  options = sorted(options, reverse=True ,key=lambda entry: entry['difficulty'])
  print(f'Available .osu maps for {choice["name"]}')
  for i in range(len(options)):
    print(f'{i:02d}) {options[i]["name"]} {options[i]["difficulty"]} {options[i]["artist"]}')
  print(f'.osu map selection(0-{len(options)-1}):')
  i = input()
  while not (int(i) < len(options)):
    print(f'Invalid input {i}, please retry(0-{len(options) - 1}): ')
    i = input()
  i = int(i)
  if i==-1:
    print("Skip")
  else:
    osu_path = os.path.join(choice['dir_path'], options[i]['filename'])
    music_path = os.path.join(choice['dir_path'], options[i]['beatmap'].general_settings.audio_file_name.strip())
    result_dir_path = os.path.join(result_dir, choice['name'])
    os.makedirs(result_dir_path)
    result_osu_path = os.path.join(result_dir_path, 'beatmap.osu')
    result_music_path = os.path.join(result_dir_path, 'audio.mp3')
    shutil.copy(osu_path, result_osu_path)
    shutil.copy(music_path, result_music_path)
    print("Files copied")

  
```
